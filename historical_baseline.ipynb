# CalCOFI Historical Baseline
Computes the 70-year average temperature and salinity (0–500 m depth) from the CalCOFI Bottle dataset.
Output: `calcofi_baseline.csv` — loaded by the app as the "Historical Normal" for Nori comparisons.

```python
import pandas as pd

DATA_PATH = (
    "CalCOFI_Database_194903-202105_csv_16October2023/"
    "CalCOFI_Database_194903-202105_csv_16October2023/"
    "194903-202105_Bottle.csv"
)
OUTPUT_PATH = "calcofi_baseline.csv"
DEPTH_MIN, DEPTH_MAX = 0, 500
```

```python
# Efficient load — only the three columns we need
# encoding='latin-1' required: file contains non-UTF-8 characters (e.g. µ symbol)
cols = ["Depthm", "T_degC", "Salnty"]
df = pd.read_csv(DATA_PATH, usecols=cols, low_memory=False, encoding="latin-1")
print(f"Loaded {len(df):,} rows")
```

```python
# Keep only upper ocean (0–500 m) and drop rows missing temp or salinity
df = df[(df["Depthm"] >= DEPTH_MIN) & (df["Depthm"] <= DEPTH_MAX)]
df = df[["T_degC", "Salnty"]].dropna()
print(f"Rows after depth filter + NaN removal: {len(df):,}")
```

```python
# Calculate baseline averages
mean_temp = df["T_degC"].mean()
mean_salt = df["Salnty"].mean()

print("=== CalCOFI 70-Year Historical Baseline (0–500 m) ===")
print(f"Mean Temperature : {mean_temp:.4f} °C")
print(f"Mean Salinity    : {mean_salt:.4f} PSU")
```

```python
# Save to CSV for the app
baseline = pd.DataFrame({
    "metric": ["mean_temperature_degC", "mean_salinity_PSU"],
    "value":  [mean_temp, mean_salt]
})
baseline.to_csv(OUTPUT_PATH, index=False)
print(f"Saved → {OUTPUT_PATH}")
baseline
```
